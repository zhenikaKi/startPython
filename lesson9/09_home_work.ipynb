```python
# Библиотека для работы с табличными данными
import pandas as pd
```

```python
# Загружаем данные
df = pd.read_csv('sample_data/california_housing_train.csv')
df.head()
```

**Задача 40:** Работать с файлом *california_housing_train.csv*, который находится в папке *sample_data*. Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population)

```python
# Средняя стоимость, где численность от 0 до 500
df.loc[ (df['population'] > 0) & (df['population'] < 500), ['population', 'median_house_value'] ]
```

**Задача 42:** Узнать какая максимальная households в зоне минимального значения population

```python
df.loc[ df['population'] == df['population'].min(), ['households'] ].max()
```
