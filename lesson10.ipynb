```python
# генерация dataFrame
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
# функция кодирования по заданным параметрам
def convert(actual_value, except_value):
  return 1 if actual_value == except_value else 0

# кодирование в one hot
data.loc[:, 'is_robot'] = data['whoAmI'].apply(lambda x: convert(x, 'robot'))
data.loc[:, 'is_human'] = data['whoAmI'].apply(lambda x: convert(x, 'human'))
data.head(20)
```
